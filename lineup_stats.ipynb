```python
import pandas as pd

off_dict = {}
def_dict = {}

num_games = 7

for game in range(1, num_games + 1):

    off_dict[f'off{game}'] = pd.read_csv(f'Lineup stats - Gm {game} Off.csv')
    def_dict[f'def{game}'] = pd.read_csv(f'Lineup stats - Gm {game} Def.csv')
```

Cumulative Lineup Stats

```python
lineup_lst = []

for game in range(1, num_games + 1):
  off_df = off_dict[f'off{game}']
  def_df = def_dict[f'def{game}']

  _lineup = pd.merge(off_df, def_df, on = 'Lineup', how = 'outer', suffixes = ('_off', '_def'))

  lineup_lst.append(_lineup)

all_games = pd.concat(lineup_lst)
```

```python
totals = all_games.groupby('Lineup').agg({
    'Poss_off': 'sum',
    'Poss_def': 'sum',
    'Points_off': 'sum',
    'Points_def': 'sum',
    'FGM_off': 'sum',
    'FGM_def': 'sum',
    'FGA_off': 'sum',
    'FGA_def': 'sum',
    '3PM_off': 'sum',
    '3PM_def': 'sum',
    '3PA_off': 'sum',
    '3PA_def': 'sum',
    'FTM_off': 'sum',
    'FTM_def': 'sum',
    'FTA_off': 'sum',
    'FTA_def': 'sum',
    'TO_off': 'sum',
    'TO_def': 'sum'}).reset_index()

totals['Plus_Minus'] = totals['Points_off'] - totals['Points_def']
totals['PPP_off'] = totals['Points_off'] / totals['Poss_off']
totals['PPP_def'] = totals['Points_def'] / totals['Poss_def']

totals.to_csv('lineup_totals.csv', index=False)
```

```python
duplicates = totals['Lineup'].duplicated(keep=False)
duplicated_lineups = totals.loc[duplicates, 'Lineup'].unique()
duplicated_lineups

# checking for duplicates
```
